```python
import pandas as pd
dataframes=[]

for file_name in os.listdir("./Farud Detection Syetem/1"):
  with open(f"./Farud Detection Syetem/1/{file_name}", "r") as f:
    dataframes.append(pd.read_csv(f))
```

```python
train_df,test_df=dataframes
```

```python
unnessary_columns_for_fraud_detection=["Unnamed: 0","cc_num","trans_num","first","last","unix_time","merchant","street"]
```

```python
train_df=train_df.drop(columns=unnessary_columns_for_fraud_detection)
test_df=test_df.drop(columns=unnessary_columns_for_fraud_detection)
```

```python
train_df["age"]=2026-pd.to_datetime(train_df["dob"]).dt.year
train_df=train_df.drop(columns="dob")
```

```python
test_df["age"]=2026-pd.to_datetime(test_df["dob"]).dt.year
test_df=test_df.drop(columns="dob")
```

```python
train_df.columns
```

```python
train_df=pd.get_dummies(train_df,columns=["gender","category"],dtype=int)
test_df=pd.get_dummies(test_df,columns=["gender","category"],dtype=int)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

le_city = LabelEncoder()
all_cities = pd.concat([train_df["city"].astype(str), test_df["city"].astype(str)]).unique()
le_city.fit(all_cities)
train_df["city"] = le_city.transform(train_df["city"].astype(str))
test_df["city"] = le_city.transform(test_df["city"].astype(str))

le_state = LabelEncoder()
all_states = pd.concat([train_df["state"].astype(str), test_df["state"].astype(str)]).unique()
le_state.fit(all_states)
train_df["state"] = le_state.transform(train_df["state"].astype(str))
test_df["state"] = le_state.transform(test_df["state"].astype(str))

le_job = LabelEncoder()
all_jobs = pd.concat([train_df["job"].astype(str), test_df["job"].astype(str)]).unique()
le_job.fit(all_jobs)
train_df["job"] = le_job.transform(train_df["job"].astype(str))
test_df["job"] = le_job.transform(test_df["job"].astype(str))
```

```python
train_df["trans_date_trans_time"]=pd.to_datetime(train_df["trans_date_trans_time"])
test_df["trans_date_trans_time"]=pd.to_datetime(test_df["trans_date_trans_time"])
```

```python
train_df["trans_year"]=train_df["trans_date_trans_time"].dt.year
test_df["trans_year"]=test_df["trans_date_trans_time"].dt.year

train_df["trans_month"]=train_df["trans_date_trans_time"].dt.month
test_df["trans_month"]=test_df["trans_date_trans_time"].dt.month

train_df["trans_day"]=train_df["trans_date_trans_time"].dt.day
test_df["trans_day"]=test_df["trans_date_trans_time"].dt.day
```

```python
train_df["trans_hour"]=train_df["trans_date_trans_time"].dt.hour
test_df["trans_hour"]=test_df["trans_date_trans_time"].dt.hour

train_df["trans_min"]=train_df["trans_date_trans_time"].dt.minute
test_df["trans_min"]=test_df["trans_date_trans_time"].dt.minute

train_df["trans_sec"]=train_df["trans_date_trans_time"].dt.second
test_df["trans_sec"]=test_df["trans_date_trans_time"].dt.second
```

```python
train_df=train_df.drop("trans_date_trans_time",axis=1)
test_df=test_df.drop("trans_date_trans_time",axis=1)
```

```python
X_train=train_df.drop("is_fraud",axis=1)
X_test=test_df.drop("is_fraud",axis=1)

Y_train=train_df["is_fraud"]
Y_test=test_df["is_fraud"]
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model=LogisticRegression()
logistic_model.fit(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model=RandomForestClassifier()
random_forest_model.fit(X_train,Y_train)
```

```python
y_train_pred_logistic=logistic_model.predict(X_train)
y_train_pred_random_forest=random_forest_model.predict(X_train)

y_tese_pred_logistic=logistic_model.predict(X_test)
y_tese_pred_random_forest=random_forest_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score

#for logistic regression model
print("="*20+"Logistic Regression Model"+"="*20)

print("Training Accuracy for Logistic Regression Model : ",accuracy_score(Y_train,y_train_pred_logistic))
print("Testing Accuracy for Logistic Regression Model : ",accuracy_score(Y_test,y_tese_pred_logistic))

print("Training F1 Score for Logistic Regression Model : ",f1_score(Y_train,y_train_pred_logistic))
print("Testing F1 Score for Logistic Regression Model : ",f1_score(Y_test,y_tese_pred_logistic))
```

```python
#for random forest  model
print("="*20+"Random Forest Classification Model"+"="*20)

print("Training Accuracy for Random Forest Classification Model : ",accuracy_score(Y_train,y_train_pred_random_forest))
print("Testing Accuracy for Random Forest Classification Model : ",accuracy_score(Y_test,y_tese_pred_random_forest))


print("Training F1 Score for Random Forest Classification Model : ",f1_score(Y_train,y_train_pred_random_forest))
print("Testing F1 Score for Random Forest Classification Model : ",f1_score(Y_test,y_tese_pred_random_forest))
```
